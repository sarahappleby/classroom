```python
import matplotlib.pyplot as plt
import numpy as np

import warnings
warnings.filterwarnings('ignore')

plt.rcParams.update({'text.usetex':True,
                     'font.family':'sans-serif'})
```

\section{Examples}

Writing figure labels in LaTeX style gives the plots a more scientific feel:

```python
x = np.arange(-10, 11, 0.1)
y = x**2

fig, ax = plt.subplots(1, 2, figsize=(10, 4))

ax[0].plot(x, y)
ax[0].set_xlabel('Lambda')

ax[1].plot(x, y)
ax[1].set_xlabel(r'$\Lambda$');
```

```python
x = np.arange(-50, 50, 1)
y = x**2

fig, ax = plt.subplots(1, 2, figsize=(10, 4))

ax[0].plot(x, x/y)
ax[0].set_xlabel(r'$x$');
ax[0].set_ylabel(r'$f = \frac{x}{y}$');

ax[1].plot(x, np.sqrt(x))
ax[1].set_xlabel(r'$x$');
ax[1].set_ylabel(r'$f = \sqrt{x}$');
```

```python
slope = np.random.uniform(low=0, high=25, size=1000)

fig, ax = plt.subplots(1, 2, figsize=(10, 4))
ax[0].hist(slope, bins=20,)
ax[0].set_xlabel(r'${\rm Slope}\ (^\circ)$');

ax[1].hist(np.log10(slope), bins=20)
ax[1].set_xlabel(r'${\rm log}\ ({\rm Slope} / ^\circ)$');
```
