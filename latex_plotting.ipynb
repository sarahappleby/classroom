## Mathtext in Matplotlib

See [here](https://matplotlib.org/stable/users/explain/text/mathtext.html) for a full reference of how to format maths from LaTeX in Matplotlib.

```python
import matplotlib.pyplot as plt
import numpy as np

import warnings
warnings.filterwarnings('ignore')

plt.rcParams.update({'text.usetex':True,
                     'font.family':'sans-serif',
                     'font.size':14})
```

```python
x = np.arange(-50, 50, 1)
y = x**2

fig, ax = plt.subplots(1, 3, figsize=(17, 4))

ax[0].plot(x, 1/x)
ax[0].set_xlabel('x')
ax[0].set_ylabel('1/x')

ax[1].plot(x, np.sqrt(x))
ax[1].set_xlabel('x')
ax[1].set_ylabel('sqrt(x)')

ax[2].plot(x, x**2)
ax[2].set_xlabel('x')
ax[2].set_ylabel('x**2');
```

```python
x = np.arange(-50, 50, 1)

fig, ax = plt.subplots(1, 3, figsize=(17, 4))

ax[0].plot(x, 1/x)
ax[0].set_xlabel(r'$x$');
ax[0].set_ylabel(r'$\frac{1}{x}$');

ax[1].plot(x, np.sqrt(x))
ax[1].set_xlabel(r'$x$');
ax[1].set_ylabel(r'$\sqrt{x}$');

ax[2].plot(x, x**2)
ax[2].set_xlabel(r'$x$');
ax[2].set_ylabel(r'$x^2$');
```

```python
slope = np.random.uniform(low=0, high=25, size=1000)

fig, ax = plt.subplots(1, 2, figsize=(10, 4))
ax[0].hist(slope, bins=20, density=True)
ax[0].set_xlabel(r'${\rm Slope}\ (^\circ)$');

ax[1].hist(np.log10(slope), bins=20, density=True)
ax[1].set_xlabel(r'${\rm log}\ ({\rm Slope} / ^\circ)$');
```
